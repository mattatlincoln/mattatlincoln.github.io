```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import patches
import matplotlib.patheffects as path_effects
%matplotlib inline
```

```python
ax = plt.axes()

arc = patches.Arc((0.0, 0.0), 0.7, 0.5, linewidth=1, fill=False, zorder=2, theta1=355, theta2=180, angle=10)
plt.gca().add_patch(arc)
arc = patches.Arc((-0.195, -0.05), 0.3, 0.15, linewidth=1, 
                fill=False, zorder=2, theta1=0, theta2=180,angle=10,
                path_effects=[path_effects.SimpleLineShadow(offset=(2, -4), linewidth=10, shadow_color='k', alpha=0.3, rho=2.0),
                    path_effects.Normal()])
plt.gca().add_patch(arc)
arc = patches.Arc((0.15, 0.0), 0.4, 0.2, linewidth=1, 
                fill=False, zorder=2, theta1=0, theta2=180,angle=10,
                path_effects=[path_effects.SimpleLineShadow(offset=(2, -4), linewidth=10, shadow_color='k', alpha=0.3, rho=2.0),
                    path_effects.Normal()])
plt.gca().add_patch(arc)

style="Simple,tail_width=1.0,head_width=4,head_length=8"
kw = dict(arrowstyle=style)

a1 = patches.FancyArrowPatch((-0.2,0.05), (0.1,0.15),connectionstyle="arc3,rad=-.2", color="k", **kw)
plt.gca().add_patch(a1)
plt.text(-0.26,0.05, r"$\mathbf{C}_n$", fontsize=16)
plt.text(0.11,0.15, r"$\mathbf{C}_{n+1}$", fontsize=16)
plt.text(-0.1,0.08, r"geodesic", fontsize=12)

a2 = patches.FancyArrowPatch((-0.2,0.05), (0.1,0.35), color='LightSalmon', **kw)
plt.gca().add_patch(a2)
plt.annotate('direction of steepest \n descent on manifold', color='LightSalmon', fontsize=12, 
             xy=(0.0, 0.25), xytext=(-0.5, 0.4),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )

a2 = patches.FancyArrowPatch((0.1,0.35), (0.1,0.15), color='blue', **kw)
plt.gca().add_patch(a2)
plt.text(0.12,0.26, r"correction", fontsize=12, color='blue')

plt.text(0.12,-0.16, r"constraint manifold for $\mathbf{C^TSC} = \mathbf{1}$", fontsize=12, color='black')

plt.axis('scaled')
plt.axis('off')

plt.savefig('OT1.svg')
```

```python
ax = plt.axes()

l1 = plt.Line2D((0.0, 0.8), (0.0, 0.4), lw=2, color="black", linestyle='-',
                path_effects=[path_effects.SimpleLineShadow(offset=(2, -4), linewidth=10, shadow_color='k', alpha=0.3, rho=2.0),
                path_effects.Normal()])
l2 = plt.Line2D((0.8, 0.0), (0.4, 0.4), lw=2, color="black", linestyle='-')
l3 = plt.Line2D((0.0, -0.8), (0.4, 0.0), lw=2, color="black", linestyle='-')
l4 = plt.Line2D((-0.8, 0.0), (0.0, 0.0), lw=2, color="black", linestyle='-',
                path_effects=[path_effects.SimpleLineShadow(offset=(2, -4), linewidth=10, shadow_color='k', alpha=0.3, rho=2.0),
                path_effects.Normal()])

for line in [l1,l2,l3,l4]:
    plt.gca().add_line(line)

#style="Simple,tail_width=1.0,head_width=4,head_length=8"
#kw = dict(arrowstyle=style)


a2 = patches.FancyArrowPatch((-0.3,0.05), (0.3,0.35), color='LightSalmon', **kw)
plt.gca().add_patch(a2)
plt.text(-0.45,0.05, r"$\mathbf{C}_n$", fontsize=16)
plt.text(0.3,0.30, r"$\mathbf{C}_{n+1}$", fontsize=16)
plt.annotate('direction of steepest \n descent on manifold', color='LightSalmon', fontsize=12, 
             xy=(0.0, 0.25), xytext=(-0.8, 0.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )

plt.text(-0.1,-0.16, r"constraint manifold for $\mathbf{X^TSC_0} = \mathbf{1}$", fontsize=12, color='black')

plt.axis('scaled')
plt.axis('off')
plt.axis([-1,1.5,-0.25,1])
plt.savefig('OT2.svg')
```
