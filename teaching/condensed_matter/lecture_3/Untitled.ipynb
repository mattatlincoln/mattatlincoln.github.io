```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
k = np.linspace(-np.pi, np.pi)
E = -10 + 3*k**2
kHalfFilled = np.linspace(-np.pi/2, np.pi/2)
EHalfFilled = -10 + 3*kHalfFilled**2
```

```python
plt.plot(k,E,linestyle=':',linewidth=3);
plt.plot(kHalfFilled,EHalfFilled, linewidth=5);
plt.xlabel(r'k');
plt.ylabel(r'E');
plt.savefig('partialFillBand.jpg')
```

```python
kMField = np.linspace(-np.pi/2,0)
kPField = np.linspace(0,+np.pi/2)
EMField = -10 +0.5 + 3*kMField**2
EPField = -10 -0.5 + 3*kPField**2

kMHalfFilledField = np.linspace(-np.pi/4,0)
EMHalfFilledField = -10 +0.5 + 3*kMHalfFilledField**2

kHalfFilledField = np.linspace(0, np.pi/4+0.17)
EHalfFilledField = -10 -0.5 + 3*kHalfFilledField**2
```

```python
plt.plot(kMField,EMField, linestyle=':',linewidth=5);
plt.plot(kPField,EPField, linestyle=':',linewidth=5);
plt.plot(kMHalfFilledField,EMHalfFilledField, linewidth=5);
plt.plot(kHalfFilledField,EHalfFilledField, linewidth=5);
plt.xlabel(r'k');
plt.ylabel(r'E');
plt.savefig('partialFillBandField.jpg')
```

```python
plt.plot(kMField,EMField, linestyle=':',linewidth=5);
plt.plot(kPField,EPField, linestyle=':',linewidth=5);
plt.plot(kMField,EMField, linestyle='-',linewidth=5);
plt.plot(kPField,EPField, linestyle='-',linewidth=5);
plt.xlabel(r'k');
plt.ylabel(r'E');
plt.savefig('FullBandField.jpg')
```
