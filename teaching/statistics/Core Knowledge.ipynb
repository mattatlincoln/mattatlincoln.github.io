### Probability spaces

<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 
8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">

$\textbf{Frequentist view of probability}$

repeated experiments ($N$ experiments) and counting the number of times something $A$ occurs ($N_A$)

$$
P(A)=p=\lim_{N \to \infty} \frac{N_A}{N}
$$
</div>

<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 
8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$\textbf{Definition:}$ the sample space is the set containing all possible outcomes of an experiment.
</div>

<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 
8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$\textbf{Definition}:$ Outcomes are the individual elementary elements of the sample space.
</div>

<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 
8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$\textbf{Definition:}$ an event is any subset, $E$, of the sample space, $S$. $E \subset A$
</div>


<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 
8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$\textbf{Axiom 1}$ - The probability that an event will come from the sample space is unity
$$P(S)=1$$
</div>
<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 
8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$\textbf{Axiom 2}$ 
$$P(A) \geq 1 \text{  for all  } A \subset S$$
</div>
<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 
8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$\textbf{Axiom 3}$
$$
P(A \cup B) = P(A) + P(B) \text{    if    } A \cap B = \emptyset
$$
</div>


<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 
8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
Probability of an event, $E$, is the sum of the probabilities of the outcomes in $E$

$$P(E)= \sum \limits_{\omega \in E} P(\omega) $$
</div>

<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 
8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$\textbf{Definition:}$ Complement of an event $E$ (with respect to the sample set $S$) is $P(\bar{E}) = 1 - P(E)$.
<br><br>

$\textbf{Definition:}$ $E_1$ and $E_2$ are mutually exclusive if $P(E_1 \cap E_2) = P\{\emptyset\} = 0$
<br><br>

$\textbf{Definition:}$ A set of events $E_1, E_2, \ldots, E_n$ of some experiment are said to be *exhaustive* if $E_1 \cup E_2 \cup E_3 \cup \ldots \cup E_n = S$.
<br><br>

If $E_1$ and $E_2$ are events of the same experiment $P(E_1 \cup E_2) = P(E_1) + P(E_2) - P(E_1 \cap E_2)$
</div>

#### Counting rules

If all outcomes are equally likely then

$$P(E)=\frac{\text{Number of outcomes in } E}{ N}$$

and we just count the number of possible outcomes. Maybe using combinatorials and permutations:

<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 
8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$\textbf{Definition:}$ We define the number of combinations of $r$ objects taken from a set of size $n$, ${n \choose r}$, for $r≤n$, by
$$
{n \choose r}=\frac{n!}{(n−r)!r!}
$$
</div>

<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 
8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$\textbf{Definition:}$ We define the number of permutations of $N$ objects  by
$$
P_N= N!
$$
</div>


For a continuous random variable

$$
P(E) = \int_{E} f(x) \mathrm{d}x,
$$


<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 
8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$\textbf{Definition}$
the conditional probability of $E$ given $F$ is 

$$
P(E \mid F) = \frac{P(E \cap F)}{P(F)}
$$
</div>

<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 
8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$\textbf{Multiplicative rule}$
$$
P(E_1 E_2 E_3 \dots E_n) = P(E_1) \frac{P(E_1 \cap E_2)}{P(E_1)} \frac{P(E_1 \cap E_2 \cap E_3)}{P(E_1 \cap P(E_2)} \cdots \frac{P(E_1 \cap E_2 \cdots \cap E_n)}{P(E_1 \cap E_2 \cdots \cap E_{n-1})}
$$
</div>

<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 
8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$\textbf{The law of total probability}$
$$
P(A) = \sum_{i=1}^n P(A \cap E_i) = \sum_{i=1}^n P(A \mid E_i) P(E_i)
$$
</div>
<br>
<br>
<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 
8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$\textbf{Bayes' Formula}$
$$
P(E_i \mid A) = \frac{P(A \cap E_i)}{P(A)} = \frac{P(A \mid E_i) P(E_i)}{\sum_{i=j}^n P(A \mid E_j) P(E_j)}
$$
</div>

## Random Variables
<br>

<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$\textbf{Definition}$  
Let $X$ be a discrete random variable that can take on only the values $x_1,x_2,x_3,\ldots,x_n$ with respective probabilities $p_X(x_1),p_X(x_2),p_X(x_3),\ldots,p(x_n)$.  
<br>
Then, if 
$$
\sum_{x \in X} p_X(x) = 1,
$$
<br>
$X$ is a *discrete random variable*.  
<br>
The function $p(x) = P\{X = x\}$ is called the probability (mass) function of the variable $X$.
</div>

$\textbf{Check list}$

- is the sample space well defined?
- are the events mutually exclusive?
- do the events cover all the sample space?
- are the probabilities of the events defined
- are values of the variable clearly assigned one-to-one to the possible events?

<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$\textbf{Cumulative distribution function}$

The cumulative distribution function of a discrete random variable, $X$,  is 

$$
F_X(a) = P\{X \leq a\} = \sum_{x \leq a} p_X(x)
$$

$\textbf{Probability density function}$ $f_X(x)$ gives the probability

$$
P\{a \leq X \leq b\} = \int_{a}^{b}f_X(x) \mathrm{d}x
$$

</div>



<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$\textbf{Cumulative distribution function for a continuous random variable, $X$}$
$$
F_X(a) = P\{X \leq a\} = \int_{-\infty}^{a}f(x) \mathrm{d}x,
$$

note that [conversely the fundamental theorem of calculus](https://en.wikipedia.org/wiki/Fundamental_theorem_of_calculus) implies that
$$
\frac{d}{da}F(a) = f(a)
$$
</div>
<br>
<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
The following will hold, exactly as for the discrete case: 

<li>$F_X(a)$ must be 0 as $a \to -\infty$ and 1 as $a \to \infty$.</li>  
<li>$F_X(a)$ must be monotonically increasing.</li>
<li>$F_X(a)$ must be [right continuous](https://www.youtube.com/watch?v=fm07adZ_WHo).</li>
</div>

when these conditions hold, $F_X$ defines a random variable $X$.

<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
To be a valid probability density function of a random variable, we must have

$$
f(t) \geq 0 \mbox{,      for all $t$}
$$

$$
\int_{-\infty}^{\infty}f(t) \text{d}t = 1
$$

</div>


Given a random variable $X$ with a probability mass function $p(x)$  we define the expectation of $X$, written as $\text{E}[X]$ as 

<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$\textbf{Definition}$

$$
\text{E}[X] = \sum_{x\in X} x \cdot p(x)
$$
</div>

The equivalent for a continuous random variable $Z$ is

<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$$
\text{E}[Z] = \int_{-\infty}^{\infty} z\cdot f(z) \mathrm{d}z
$$
</div>

We see that the relationship is very close between discrete and continuous cases - we replace a sum with an integral.

<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
We also call $\text{E}[X]$ the mean of $X$ and write it as $\mu_X$
</div>

<br>

<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$\textbf{Definition}$

Let $g(X)$ be any function of a discrete random variable $X$. 

Then

$$
\text{E}[g(X)] = \sum_{x \in X}g(x) \cdot p(x)
$$

</div>
<br>

<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">

Let $h(Z)$ be any function of  the continuous random variable $Z$

$$
\text{E}[h(Z)] = \int_{-\infty}^{\infty} h(z)\cdot f(z) \mathrm{d}z
$$
</div>
<br><br>
<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$\textbf{Variance and Standard Deviation}$

If $X$ is a random variable with mean $\mu$, then the variance of $X$, denoted by $\text{Var}(X)$ or $\sigma_X^2$, is defined by

$$
\sigma_X^2 = \text{Var}(X) = \text{E}[(X-\mu)^2]
$$

this can also be written

$$
\sigma_X^2 = \text{Var}(X) = \text{E}[X^2] - (\text{E}[X])^2 
$$

the standard deviation $\sigma_X$ is 

$$
\sigma_X = \sqrt{\text{Var}(X)} = \sqrt{\text{E}[X^2] - (\text{E}[X])^2 } = \sqrt{\text{E}[(X-\mu)^2]}
$$

</div>


# Joint random variables.

We defined two random variables - $X$ and $Y$. What about if it was important to consider the *joint* probability mass function of $X$ and $Y$. This is written as 


<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
$$P\{X=x,Y=y\} = p(x,y)$$

for discrete random variables, or

$$P\{X=x,Y=y\} = f(x,y)dx dy$$

for continuous ones.
<br><br>
$\textbf{Marginal distributions}$  

$$
P\{X=x,Y=y\} = p(x,y) = P(X=x \cap Y=y).
$$

We then have using the law of total probability

$$
P(A) = \sum_{i=1}^n P(A \cap E_i) = \sum_{i=1}^n P(A \mid E_i) P(E_i)
$$

we can immediately see, specialising to our case where the $E_i$ are the values of $y$

$$
P\{X=x\} = \sum_{\text{all $y$}} P(X=x \cap Y=y) = \sum_{\text{all $y$}}P(X=x \mid Y=y) P(Y=y).
$$

and equally 

$$
P\{Y=y\} = \sum_{\text{all $x$}} P(Y=y \cap X=x) = \sum_{\text{all $x$}}P(Y=y \mid X=x) P(X=x).
$$.

The covariance of two variables can be calculated using alternative expressions

$$
\begin{align}
\text{Cov}(X,Y) & = \text{E}\Big[(X - \text{E}[X])\cdot(Y - \text{E}[Y])\Big] \\
                & = \text{E}\Big[XY - \text{E}[X]Y - X\text{E}[Y] + \text{E}[X]\text{E}[Y]\Big] \\
                & = \text{E}\Big[XY] - \text{E}[X]\text{E}[Y] - \text{E}[X]\text{E}[Y] + \text{E}[X]\text{E}[Y]\Big]\\ 
                & = \text{E}[XY] - \text{E}[X]\text{E}[Y]
\end{align}
$$


in the special case that $X=Y$,

$$
\begin{align}
\text{Cov}(X,X) & = \text{E} \Big[(X - \text{E}[X])(X - \text{E}[X]) \Big] \\
                & = \text{E}\Big[(X - \text{E}[X])^2\Big] \\
                & = \text{Var}(X)
\end{align}
$$
</div>


## Measures of central tendency and variation

<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">
<br>
<li>median of a discrete set of data is the middle value of the distribution if values are listed in order, for an odd number of values. Or the mean of the two central values, for even numbers of data points.</li>
<br>
<li> median of a discrete random variable $Z$ is the value $z$ such that $P(Z \leq z) \geq 0.5$ and $P(Z \geq z) \geq 0.5$.</li>
<br>
<li> median of a continuous vandom variable $X$ is the value $x$ such that $F(x) = \frac{1}{2}$.</li>
</div>

Quantiles are cutpoints dividing a set of observations into equal sized groups.

Actually, we've already met one - the median. This cuts the set of observations/values of the distribution in 2.

For a continuous distribution life is actually easier, the median is just the value at which the cumulative distribution function of a variable reaches $\frac{1}{2}$.

<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">

<li>median of a discrete set of data is the middle value of the distribution if values are listed in order, for an odd number of values. Or the mean of the two central values, for even numbers of data points.</li>
<li>the median of a discrete random variable $Z$ is the value $z$ such that $P(Z \leq z) \geq 0.5$ and $P(Z \geq z) \geq 0.5$.</li>
<li>median of a continuous vandom variable $X$ is the value $x$ such that $F(x) = \frac{1}{2}$.</li>
</div>

<br>

<div style="background-color:Gold; margin-left: 20px; margin-right: 20px; padding-bottom: 8px; padding-left: 8px; padding-right: 8px; padding-top: 8px; border-radius: 25px;">

For a general quantile we just replace $\frac{1}{2}$ with $a/n$ where $(n-1)$ is number of cuts we will make in the data, and $0 < a < n$ gives the $a^{th}$ $n$-tile.
</div>
